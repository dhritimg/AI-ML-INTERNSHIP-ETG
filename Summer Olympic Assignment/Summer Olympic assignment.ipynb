# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
df = pd.read_csv("summer.csv")
city_count = df.City.unique()
print("The number of Cities where Olympics were held are :",len(city_count))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
Gold_data = df[df.Medal == "Gold"]
Sport_names = Gold_data['Sport']
Gold_medalCounts = Sport_names.value_counts()
print("The Sport with higest number of Gold medals won so far is :",Gold_medalCounts[0])
pd.Series(Gold_medalCounts.head()).plot(title= "Top 5 sports with higest number of Gold medals",kind = 'bar',figsize = (15,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
sport_names = df['Sport']
medal_counts = sport_names.value_counts()
print("The Sport with higest number of medals won so far is :",medal_counts.index[0])
pd.Series(medal_counts.head()).plot(title= "Top 5 sports with higest number of Medals",kind = 'bar',figsize = (15,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
Player_count = []

for name in df["Athlete"].unique():
    Player_count.append([name , len(df[df['Athlete']  == name])])
Player_count.sort(reverse = True ,key = lambda x: x[1])
pd.DataFrame(Player_count,columns = ['Athlete Name', 'Number of Medals won']).head().plot(y = 'Number of Medals won', x ='Athlete Name', kind = 'bar', figsize = (15,5))
print("The player with higest medals won is :",Player_count[0][0])
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
Gold_data = df[df.Medal == "Gold"]
Player_count_gold = []

for name in Gold_data["Athlete"].unique():
    Player_count_gold.append([name , len(Gold_data[Gold_data['Athlete']  == name])])
    
    
Player_count_gold.sort(reverse = True ,key = lambda x: x[1])
pd.DataFrame(Player_count_gold,columns = ['Athlete Name', 'Number of Gold Medals won']).head().plot(x = 'Athlete Name', y = 'Number of Gold Medals won', kind = 'bar', figsize = (15,5))
print("The player with higest Gold medals won is :",Player_count_gold[0][0])
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
data = df.sort_values(by=['Year'])
dat = data.loc[data['Country'] == 'IND']
da = dat.loc[dat.Medal == "Gold"]
print ("The year when India won it's first gold medal was :",da.iloc[0]["Year"])
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
Event_data = df['Event']
Event_data = Event_data.value_counts().head()
pd.Series(Event_data).plot(title= "Top 5 Events popular by number of players",kind = 'bar',figsize = (15,5))
print("The Event Which is most popular by Number of players is :",Event_data.index[0])
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
Women_data = df[df["Gender"] == "Women"]
Women_data =Women_data[Women_data.Medal == "Gold"]
Sport_data = Women_data['Sport']
Sport_data = Sport_data.value_counts().head()
pd.Series(Sport_data).plot(title= "Top 5 sports with higest number of Female Gold medalists",kind = 'bar',figsize = (15,5))
print("The Sport Which is most popular by the number of Female Gold Medalists is :",Sport_data.index[0])
```
