#   Reading from Inventory

```python
import json
fd = open("inventory.json",'r')
r = fd.read()
fd.close()
inventory = json.loads(r)
sale = {}
C = 0
```

# Adding New Item to The Inventory

```python
product_id = input("Enter product ID:")
product_name = input("Enter the Product Name:")
price = int(input("Enter the price:"))
quant = int(input("Enter quantity:"))
C_mark = input("Enter the color mark(Red or Green) of the product:")
Expdate = input("Enter expiry date of the product in YYYY-MM-DD format:")

inventory[product_id] = {"Product_Name": product_name, 'Price': price, 'Quant_Avl': quant, "C_Mark": C_mark, "ExpDate": Expdate}

js = json.dumps(inventory)

fd = open("inventory.json",'w')
fd.write(js)
fd.close()
```

# To Delete a product from the inventory

```python
product_id = input("Enter product ID to be deleted from the inventory:")
if product_id in inventory.keys():
    del inventory[product_id]
    print("The product has been deleted from the inventory")
else:
    print("There is no product with the product ID entered in the Inventory")

js = json.dumps(inventory)

fd = open("inventory.json",'w')
fd.write(js)
fd.close()
```
